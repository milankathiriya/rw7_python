##### Merge and join data

```python
# Creating first DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
})

# Creating second DataFrame
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Salary': [50000, 60000, 70000, 80000]
})
```

```python
df1
```

```python
df2
```

```python
# merge() & join()
```

```python
pd.merge(df1, df2, on='ID', how='inner')
```

```python
pd.merge(df1, df2, on='ID', how='left')
```

```python
pd.merge(df1, df2, on='ID', how='right')
```

```python
pd.merge(df1, df2, on='ID', how='outer')
```

```python
pd.merge(df1, df2)
```

```python
df1.merge(df2)
```

```python
# join()
```

```python
df3 = df1.set_index('ID')
df4 = df2.set_index('ID')
```

```python
df3
```

```python
df4
```

```python
df3.join(df4)
```

```python
df3.join(df4, how='left')
```

```python
df3.join(df4, how='inner')
```

```python
df3.join(df4, how='right')
```

```python
df3.join(df4, how='outer')
```

##### Visualize data

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Age': [25, 30, 35, 40, 29],
    'Salary': [50000, 60000, 70000, 80000, 55000]
}

df = pd.DataFrame(data)

df
```

```python
import matplotlib

df.plot(kind="bar", x="Name", y="Salary", title="Salary by Name", xlabel="Name", ylabel="Salary")
```

```python
df.plot(kind="line", x="Name", y="Salary", title="Salary by Name", xlabel="Name", ylabel="Salary", marker="o")
```

```python
df.plot(kind="pie", y="Salary", title="Salary by Name", labels=df["Name"], startangle=90, autopct='%0.2f%%')
```
